```python
from keras.models import load_model  # TensorFlow is required for Keras to work
import cv2  # Install opencv-python
import numpy as np
from PIL import ImageFont, ImageDraw, Image  # 載入 PIL 相關函式庫
import random
import time
import time

start = time.time()
print("start time")
print(start)
end = time.time()
print("end - start")
print(end - start)

#fontpath = 'NotoSansTC-Regular.ttf'          # 設定字型路徑
fontpath = 'NotoSansTC-VariableFont_wght.ttf'
# Disable scientific notation for clarity
np.set_printoptions(suppress=True)

question = ""
repeat = False
wait = False
paper = "paper"
scissor = "scissor"
stone = "stone"

# Load the model
model = load_model("keras_Model.h5", compile=False)

def randomQ():
    global last_random_number
    while True:
        random_number = random.randint(0, 2)
        if last_random_number != random_number:
            last_random_number = random_number
            break   
    ranQ = "";
    if random_number == 0:
        ranQ = paper
    if random_number == 1:
        ranQ = scissor
    if random_number == 2:
        ranQ = stone
    return ranQ

def text(text1,text2):                               # 建立顯示文字的函式
    global show_img                           # 設定 img 為全域變數
    org = (0,50)                              # 文字位置
    font = ImageFont.truetype(fontpath, 50)   # 設定字型與文字大小
    imgPil = Image.fromarray(show_img)        # 將 img 轉換成 PIL 影像
    draw = ImageDraw.Draw(imgPil)             # 準備開始畫畫
    draw.text((0, 0), text1, fill=(255, 255, 255), font=font)  # 寫入文字
    draw.text((0, 150), text2, fill=(255, 255, 255), font=font)  # 寫入文字
    show_img = np.array(imgPil)


cap = cv2.VideoCapture(0)
if not cap.isOpened():
    print("Cannot open camera")
    exit()
while True:
    if not repeat:
        question = randomQ()
        print("------------------------------------------------------")
        print(question)
        print("------------------------------------------------------")
        repeat = True
    
    ret, frame = cap.read()
    if not ret:
        print("Cannot receive frame")
        break
    img = cv2.resize(frame , (398, 224))
    show_img = img[0:224, 80:304]
    img = np.asarray(show_img, dtype=np.float32).reshape(1, 224, 224, 3)
    img = (img / 127.5) - 1
    if wait:
        if(end - start > 2.0):
            wait = False
        elif (end - start > 1) and (end - start < 2.0):
            end = time.time()
            text("Next Question","Ready !!")
        else:
            end = time.time()
            text("Correct","Good")
    else:
        text(question,"Beat him")
        prediction = model.predict(img)
        index = np.argmax(prediction)
        print(index)

    if (index == 0 and question == stone) or (index == 1 and question == paper) or (index == 2 and question == scissor):
        repeat = False # 重新設定題目
        wait = True # 等待正確畫面顯示
        start = time.time() # 紀錄時間
        
    cv2.imshow("Webcam Image", show_img)
    if cv2.waitKey(1) == ord('q'):
        break    # 按下 q 鍵停止
cap.release()
cv2.destroyAllWindows()
```
